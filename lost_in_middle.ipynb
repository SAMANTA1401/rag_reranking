```python
from langchain.vectorstores import Chroma
import chromadb
```

```python

import os
os.getcwd()
```

```python

DB_DIR = os.path.join(os.getcwd(),"DB/chroma")
```

```python
DB_DIR
```

```python
client_settings = chromadb.config.Settings(
    is_persistent=True,
    persist_directory=DB_DIR,
    anonymized_telemetry=False,
)
```

```python
# is_persistent=True: Enables persistent storage, allowing the database to store data across sessions.
# persist_directory=DB_DIR: Specifies the directory where the persistent data will be stored. In this case, it's set to DB_DIR.
# anonymized_telemetry=False: Disables anonymized telemetry reporting. If set to True, anonymous usage data would be sent to the Chroma developers.
```

```python
vectorstore1 = Chroma.from_documents(text,
                                    hf_embeddings,
                                    client_settings=client_settings,
                                    collection_name="dataready",
                                    collection_metadata={"hnsw":"cosine"},
                                    persist_directory="/store/dataready")
```

```python
vectorstore1
```

```python

vectorstore2 = Chroma.from_documents(text2,
                                    hf_embeddings,
                                    client_settings=client_settings,
                                    collection_name="datadriven",
                                    collection_metadata={"hnsw":"cosine"},
                                    persist_directory="/store/datadriven")
```

```python
vectorstore2
```

```python
retriever1= vectorstore1.as_retriever(search_type="mmr",search_kwargs={"k": 5, "include_metadata": True})
     
```

```python
# text2: The list of text documents to be indexed.
# hf_embeddings: The Hugging Face embeddings model used to generate vector embeddings for the text documents.
# client_settings=client_settings: The client settings configuration, which includes options for persistent storage, telemetry, and more.
# collection_name="datadriven": The name of the collection to be created in the Chroma database.
# collection_metadata={"hnsw":"cosine"}: The metadata for the collection, which specifies the HNSW (Hierarchical Navigable Small World) graph algorithm and the cosine similarity metric.
# persist_directory="/store/datadriven": The directory where the Chroma database will be persisted.
```

```python
# search_type="mmr": Specifies the search algorithm to use. In this case, it's set to "mmr", which stands for Maximal Marginal Relevance.
# search_kwargs={"k": 5, "include_metadata": True}: Passes additional keyword arguments to the search algorithm. Here, it specifies:
# k=5: The number of top results to return.
# include_metadata=True: Includes metadata in the search results.
# What the Code Does:
# This code creates a retriever object (retriever1) from the vectorstore1 instance, configured to use the MMR search algorithm with the specified parameters.
# MMR Search Algorithm:
# The MMR algorithm is designed to retrieve documents that are not only relevant to the query but also diverse and non-redundant. It does this by selecting documents that maximize the marginal relevance, which is the relevance of the document given the previously selected documents.
```

```python
retriever1
```

```python
retriever2 = vectorstore2.as_retriever(search_type="mmr",search_kwargs={"k": 5, "include_metadata": True})
```

```python
from langchain.document_transformers import (
    EmbeddingsClusteringFilter,
    EmbeddingsRedundantFilter,
)
from langchain.retrievers.document_compressors import DocumentCompressorPipeline
from langchain.retrievers import ContextualCompressionRetriever
from langchain.document_transformers import LongContextReorder
```

check redundant or duplicate

```python
from re import search
filter = EmbeddingsRedundantFilter(embeddings=hf_embeddings)
```

```python
filter
```

```python
reordering = LongContextReorder()
```

```python
reordering
```

```python
# This class takes into account the context of the entire document, rather than just the individual passages or snippets, to reorder the results. This can help improve the relevance and coherence of the results.
```

```python
pipeline = DocumentCompressorPipeline(transformers=[filter, reordering])
compression_retriever_reordered = ContextualCompressionRetriever(
    base_compressor=pipeline, base_retriever=lotr,search_kwargs={"k": 3, "include_metadata": True}
)
```

```python
pipeline
```

```python
compression_retriever_reordered
```

```python
from langchain.llms import LlamaCpp
```

```python

# Llama CPP is a C++ implementation of the LLaMA (Large Language Model Meta AI) model. It's a high-performance, open-source library that allows developers to integrate LLaMA's language processing capabilities into their applications.
# Llama CPP is designed to provide:
# Fast and efficient inference
# Low latency
# High throughput
```

```python
os.cpu_count() # octa core
```

```python

llms = LlamaCpp(streaming=True,
                model_path="model/zephyr-7b-beta.Q4_K_M.gguf",
                max_tokens = 1500,
                temperature=0.75,
                top_p=1, # one doc top rank
                gpu_layers=0,
                stream=True,
                verbose=True,
                n_threads = int(os.cpu_count()/2),
                n_ctx=4096)
```

```python

from langchain.chains import RetrievalQA
```

```python
qa = RetrievalQA.from_chain_type(
    llm=llms,
    chain_type="stuff",
    retriever = compression_retriever_reordered,
)
```

```python
qa
```

```python
query = " what is data driven approach ?"
```

```python
results = qa(query)
```
